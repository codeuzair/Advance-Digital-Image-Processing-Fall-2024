```python
import numpy as np;
import cv2;
import pandas;
import matplotlib.pyplot as plt;
```

# Image Sampling

### Down Sampling

```python
img = cv2.imread(r'Images\rose.jpg', cv2.IMREAD_GRAYSCALE);
```

```python
cv2.imshow("Rose", img);
cv2.waitKey(0);
cv2.destroyAllWindows();
```

```python
img.size
```

```python
img.shape
```

```python
img1 = img[0:1200:2, 0:1600:2];
cv2.imshow("Rose sampled1", img1);
cv2.waitKey(0);
cv2.destroyAllWindows();
```

```python
img1.shape
```

```python
img2 = img1[0::2, 0::2];
cv2.imshow("Rose sampled2", img2);

img3 = img2[0::2, 0::2];
cv2.imshow("Rose sampled3", img3);

img4 = img3[0::2, 0::2];
cv2.imshow("Rose sampled4", img4);

cv2.waitKey(0);
cv2.destroyAllWindows();
```

```python
fig = plt.figure(figsize=(10,7));


fig.add_subplot(2,2,1);
plt.imshow(img1, cmap='gray');
plt.title("Image sampled from origional Image");
plt.axis('off');

fig.add_subplot(2,2,2);
plt.imshow(img2, cmap='gray');
plt.title("Image sampled from image 1");
plt.axis('off');

fig.add_subplot(2,2,3);
plt.imshow(img3, cmap='gray');
plt.title("Image sampled from image 2");
plt.axis('off');

fig.add_subplot(2,2,4);
plt.imshow(img4, cmap='gray');
plt.title("Image sampled from image 3");
plt.axis('off');
```

### Up Sampling

```python
plt.imshow(img4, cmap='gray');
plt.title("Image 4");
plt.axis('off')
plt.show();
```

```python
img3.shape
```

```python
upsample_img4 = np.zeros((150,200), dtype=np.uint8);
```

```python
upsample_img4[::2, ::2] = img4; 
```

```python
plt.imshow(upsample_img4, cmap='gray');
plt.title("Upsampled Image 4");
plt.axis('off');
plt.show();
```

```python
upsample_img4[1::2, 1::2] = img4; 
plt.imshow(upsample_img4, cmap='gray');
plt.title("Upsampled Image 4");
plt.axis('off');
plt.show();
```

```python
upsample_img4[::2, 1::2] = img4; 
plt.imshow(upsample_img4, cmap='gray');
plt.title("Upsampled Image 4");
plt.axis('off');
plt.show();
```

```python
upsample_img4[1::2, ::2] = img4; 
plt.imshow(upsample_img4, cmap='gray');
plt.title("Upsampled Image 4");
plt.axis('off');
plt.show();
```
