```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img=cv2.imread('Fig0308(a)(fractured_spine).tif', 0)
imgp=(img**0.3)

```

```python
cv2.imshow('Original', img)
cv2.imshow('After change', cv2.normalize(imgp, None, 0, 1, cv2.NORM_MINMAX))
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
fig, ax=plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(img, cmap='gray')
ax[1].imshow(imgp, cmap='gray')
```

```python
img=cv2.imread('Fig0316(2)(2nd_from_top).tif', 0)

x, y=img.shape

img2=np.zeros((x,y), dtype='uint8')

r1=96
r2=150
s1=42
s2=224
a1=1.5
a2=2.5
L=256
a3=1.2
for i in range(0, 500):
    for j in range(0, 500):
        if img[i, j]>=0 and img[i, j]<r1:
            img2[i, j]=(img[i, j]*a1).astype("B")
        elif img[i, j]>=r1 and img[i, j]< r2:
            img2[i, j]=a2*(img[i, j]-r1)+s1
        elif img[i, j]>=r2 and img[i, j]< L-1:
            img2[i, j]=a3*(img[i, j]-r2)+s2

    
```

```python


cv2.imshow('Original', img)
cv2.imshow('After change', img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img=cv2.imread('Fig0316(2)(2nd_from_top).tif', 0)

x, y=img.shape

img2=np.zeros((x,y), dtype='uint8')

for i in range(0, 500):
    for j in range(0, 500):
        if img[i, j]<=100:
            img2[i, j]=0
        elif img[i, j]>100:
            img2[i, j]=255
fig, ax=plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(img, cmap='gray')
ax[1].imshow(img2, cmap='gray')
```

```python
img=cv2.imread('Fig0316(2)(2nd_from_top).tif', 0)
img2 = np.where(img>=100, 255, 0).astype('B')
plt.imshow(img2, cmap="gray")
```

```python
img=cv2.imread('Fig0312(a)(kidney).tif', 0)
A=150
B=230
cond1=img>=A
cond2=img<=B
img2 = np.where(cond1 & cond2, img, 0).astype("B")
plt.imshow(img2, cmap="gray")
```
