```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('Fig0304(a)(breast_digital_Xray).tif', 0)
g=256-1-img
fig, ax=plt.subplots(1, 2,figsize=(8,8))
ax[0].imshow(img, cmap='gray')
ax[0].set_title("Original Image");
ax[0].axis('off');
ax[1].imshow(g, cmap='gray')
ax[1].set_title("Original Image");
ax[1].axis('off');
```

```python
img=cv2.imread('scaling_img.png', 0)
g=(img*2.6).astype(np.uint8)

g=cv2.normalize(g, None, 0, 255, cv2.NORM_MINMAX)

fig, ax=plt.subplots(1, 2,figsize=(8,8))
ax[0].imshow(img, cmap='gray')
ax[0].set_title("Original Image");
ax[0].axis('off');
ax[1].imshow(g, cmap='gray')
ax[1].set_title("After Transformation");
ax[1].axis('off');
```

```python
img=cv2.imread('Fig0305(a)(DFT_no_log).tif', 0)

g=np.log(1+img)
fig, ax=plt.subplots(1, 2,figsize=(8,8))
ax[0].imshow(img, cmap='gray')
ax[0].set_title("Original Image");
ax[0].axis('off');
ax[1].imshow(g, cmap='gray')
ax[1].set_title("After Transformation");
ax[1].axis('off');
```

```python
img=cv2.imread('Fig0307(a)(intensity_ramp).tif', 0)
c=1
gm=2.5
imggc2=c*img**gm
imggc4=c*imggc2**0.4
imggc3=c*img**0.4

fig, ax=plt.subplots(2, 2, figsize=(10, 10))
ax[0, 0].imshow(img, cmap='gray')
ax[0, 0].set_title('Original')

ax[0, 1].imshow(imggc2, cmap='gray')
ax[0, 1].set_title('Gamma 2.5')

ax[1, 0].imshow(imggc3, cmap='gray')
ax[1, 0].set_title('Gamma 0.4')

ax[1, 1].imshow(imggc4, cmap='gray')
ax[1, 1].set_title('Gamma 0.4')
```

```python
img=cv2.imread("Fig0316(2)(2nd_from_top).tif", 0)
x, y=img.shape
imgout=np.zeros((x, y), dtype="B")

r1=92
r2=160
s1=42
s2=224
a1=0.75
a2=3.75
a3=1.25
L=256



for i in range(0, x):
    for j in range(0, y):
        if img[i, j]>=0 and img[i, j]< r1:
            imgout[i, j]=a1*img[i, j]
        elif img[i, j]>=r1 and img[i, j]< r2:
            imgout[i, j]=a2*(img[i, j]-r1)+s1
        elif img[i, j]>=r2 and img[i, j]< L-1:
            imgout=a3*(img[i, j]-r2)+s2

cv2.imshow('Original Imag', img)
cv2.imshow('Output image', imgout)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread("Fig0316(2)(2nd_from_top).tif", 0)
x, y=img.shape
imgout=np.zeros((x, y), dtype="B")
imgout=np.where(img<110, 0, 255)

fig, ax=plt.subplots(1, 2, figsize=(15, 10))
ax[0].imshow(img, cmap='gray')
ax[0].set_title('Original')

ax[1].imshow(imgout, cmap='gray')
ax[1].set_title('Gamma 0.6')
```
