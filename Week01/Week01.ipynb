## Environments
---
- Environments in conda are self-contained, isolated spaces where you can install specific versions of software packages, including dependencies, libraries, and Python versions. 
- This isolation helps avoid conflicts between package versions and ensures that your projects have the exact libraries and tools they need.

## Environment Management
---
- Create environment with Python version &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;     `conda create -n ENVNAME python=3.10` 

- Activate environment (do this before installing packages)   &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;   `conda activate ENVNAME` 

- Deactivate environment        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;     `conda deactivate`

- List installed packages        &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;      `conda list`

- List all environments and locations     &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;     `conda info --envs`

- Delete environment by name       &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;      `conda remove -n ENVNAME --all`

- Install packages in environment     &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;     `conda install -n ENVNAME PKGNAME`

- Install packages in the activated environment     &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;     `conda install  PKGNAME`
  &nbsp;&nbsp; e.g. &nbsp;&nbsp;    Install OpenCV  &nbsp;&nbsp;  `conda install opencv` <br><br>

### Image Slicing

### Pixel Replication

```python
img = cv2.imread("images\lenna.jpg", 0);

row_rep_img = np.repeat(img, 2, axis = 0) # row wise pixels replacation
col_rep_img = np.repeat(img, 2, axis = 1) # column wise pixels replacation
replicated_img = np.repeat(row_rep_img, 2, axis=1);   # row and column size pixels replacation

cv2.imshow("Original Image", img);
cv2.imshow("Row Replicated Image", row_rep_img);
cv2.imshow("Column Replicated Image", col_rep_img);
cv2.imshow("Row/Column Replicated Image", replicated_img);
cv2.waitKey(0);
cv2.destroyAllWindows()
```

```python
print("Shape of Original Image : ", img.shape);
print("Shape of Row Replicated Image : ", row_rep_img.shape);
print("Shape of Col Replicated Image : ", col_rep_img.shape);
print("Shape of Row/Col Replicated Image : ", replicated_img.shape);
```

### Image resizing

```python
img = cv2.imread('images\Fig0219(rose1024).tif', 0)

img512 = img[::2, ::2];
img256 = img512[::2, ::2];
img128 = img256[::2, ::2];
img64 = img128[::2, ::2];
img32 = img64[::2, ::2];
            
# create figure
fig, axs = plt.subplots(2, 3, figsize=(10, 7))
  
# showing images
axs[0,0].imshow(img,cmap = 'gray')
axs[0,0].axis('off')
axs[0,0].set_title('Original image')

axs[0,1].imshow(img512, cmap = 'gray')
axs[0,1].axis('off')
axs[0,1].set_title('result512')

axs[0,2].imshow(img256,cmap = 'gray')
axs[0,2].axis('off')
axs[0,2].set_title('result256')

axs[1,0].imshow(img128,cmap = 'gray')
axs[1,0].axis('off')
axs[1,0].set_title('result128')

axs[1,1].imshow(img64,cmap = 'gray')
axs[1,1].axis('off')
axs[1,1].set_title('result64')

axs[1,2].imshow(img32,cmap = 'gray')
axs[1,2].axis('off')
axs[1,2].set_title('result32')

plt.tight_layout();
plt.show()
```

### Image Quantization
-  Redusing the intensity resolutin of image i.e. reducing number of intensity levels

```python
img = cv2.imread('images/Fig0219(rose1024).tif', 0)

# Convert the image to different bit depths
result_7_bit = (img // 2).astype('uint8')*2;
result_6_bit = (img // 4).astype('uint8')*4;
result_5_bit = (img // 8).astype('uint8')*8;
result_4_bit = (img // 16).astype('uint8')*16;
result_3_bit = (img // 32).astype('uint8')*32;
result_2_bit = (img // 64).astype('uint8')*64;
result_1_bit = (img // 128).astype('uint8')*255;

# Display the images
cv2.imshow('7-bit', result_7_bit)
cv2.imshow('6-bit', result_6_bit)
cv2.imshow('5-bit', result_5_bit)
cv2.imshow('4-bit', result_4_bit)
cv2.imshow('3-bit', result_3_bit)
cv2.imshow('2-bit', result_2_bit)
cv2.imshow('1-bit', result_1_bit)
cv2.imshow('Original', img)

# Wait for a key press and close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```
